```python
import numpy as np, faiss, json
from sentence_transformers import SentenceTransformer

MODEL_NAME = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
EMB_OUT = "../data/embeddings.npy"
INDEX_OUT = "../data/faiss_index.ivf"
META_OUT = "../data/embeddings_meta.json"

model = SentenceTransformer(MODEL_NAME)
index = faiss.read_index(INDEX_OUT)
meta = json.load(open(META_OUT,"r",encoding="utf-8"))
embs = np.load(EMB_OUT)

def search(q, k=5):
    v = model.encode([q], convert_to_numpy=True)
    faiss.normalize_L2(v)
    D,I = index.search(v, k)
    results = []
    for score, idx in zip(D[0], I[0]):
        results.append({"score": float(score), "meta": meta[idx]})
    return results

print(search("địa điểm du lịch ở quy nhơn", k=5))
```
