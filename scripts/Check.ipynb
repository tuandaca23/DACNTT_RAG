### Ktra embedding meta

```python
# check_alignment.py
import json, numpy as np, os
EMB = "../data/embeddings.npy"
META = "../data/embeddings_meta.json"
CHUNKS = "../data/pdf_chunks_cleaned_for_embed_v3.json"

emb_shape = np.load(EMB).shape if os.path.exists(EMB) else None
meta = json.load(open(META,"r",encoding="utf-8")) if os.path.exists(META) else []
chunks = json.load(open(CHUNKS,"r",encoding="utf-8")) if os.path.exists(CHUNKS) else []

print("n_emb:", emb_shape[0] if emb_shape else 0)
print("emb_dim:", emb_shape[1] if emb_shape else None)
print("n_meta:", len(meta))
print("n_chunks:", len(chunks))
print("\nmeta sample head:", meta[:3])
print("\nchunks sample head (index 0..2):")
for i,c in enumerate(chunks[:3]):
    print(i, c.get("id"), c.get("char_len"))
# quick check: are meta entries ordered same as chunks order?
same_order = all(meta[i].get("chunk_id") == chunks[i].get("id") for i in range(min(len(meta), len(chunks))))
print("\nMeta chunk_id equals chunks[i].id for all i? ->", same_order)
```
